## Скелет декоратора

```python
from typing import Callable


def empty_deco(func: Callable):
    def wrapper():
        res = func()
        return res
    return wrapper
```

```python
@empty_deco
def my_func():
    return 124
```

```python
print(my_func())
```

## Декоратор для подсчет времени

```python
from typing import Callable
import time


def timer_deco(func: Callable):
    def wrapper():
        start = time.time()
        res = func()
        end = time.time()
        print(f"Функция исполнения заняла {end - start}")
        return res
    return wrapper
```

```python
@timer_deco
def my_func():
    time.sleep(2)
    return 124

my_func()
```

## Декоратор без параметра

```python
from typing import Callable


def param_timer_deco(func: Callable):
    def wrapper(*args, **kwargs):
        start = time.time()
        res = func(*args, **kwargs)
        end = time.time()
        print(f"Функция исполнения заняла {end - start}")
        return res
    return wrapper
```

```python
@param_timer_deco
def my_func(sleep_time: int):
    time.sleep(sleep_time)
    return 124


my_func(3)
```

## Декоратор с параметром

```python
from typing import Callable


def limit_calls_0(func: Callable):
    def wrapper(*args, **kwargs):
        start = time.time()
        res = func(*args, **kwargs)
        end = time.time()
        print(f"Функция исполнения заняла {end - start}")
        return res
    return wrapper


def limit_calls(limit: int):
    def wrapper(func: Callable):
        def inner(*args, **kwargs):
            nonlocal limit
            if limit == 0:
                print("Нельзя вызвать функцию")
                return

            res = func(*args, **kwargs)
            limit -= 1
            return res
        
        return inner
    return wrapper
```

```python
@limit_calls
def my_func(sleep_time: int):
    time.sleep(sleep_time)
    return 124
```

```python
my_func()
```
